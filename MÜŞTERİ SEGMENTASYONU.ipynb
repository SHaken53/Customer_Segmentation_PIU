```python
def SuperInfo(df, dropna=False):
    """
    Returns a dataframe consisting of datatypes, nuniques, #s of nulls head(1), most frequent item and its frequncy,
    where the column names are indices.
    """
    dt=pd.DataFrame(df.dtypes, columns=["Type"])
    dn=pd.DataFrame(df.nunique(), columns=["Nunique"])
    nonnull=pd.DataFrame(df.isnull().sum(), columns=["#of Missing"])
    firstT=df.head(1).T.rename(columns={0:"First"})
    MostFreqI=pd.DataFrame([df[x].value_counts().head(1).index[0] for x in df.columns], columns=["MostFreqItem"],index=df.columns)
    MostFreqC=pd.DataFrame([df[x].value_counts().head(1).values[0] for x in df.columns], columns=["MostFreqCount"],index=df.columns)
    return pd.concat([dt,dn,nonnull,MostFreqI,MostFreqC,firstT],axis=1)
```

# Cluster Analysis

Cluster analizde iki farklı yöntem deneyeceğiz ve bunları karşılaştıracağız.

Birincisi **K-Means algoritması** diğeri ise **hiyerarşik kümeleme**. 


**KMeans Clustering Teorisi**

- Amaç gözlemleri birbirlerine olan benzerliklerine göre kümelerine ayırmaktır.

- Kaç küme oluşturacağını baştan söyleyeceğiz.

- Öyle bir k değeri belirlemeliyim ki veri setimin içerisindeki kümelenme değişimini minumum hata ile bana ortaya koymuş olsun.

- Minumum hata; kendi içindeki küme içi hataların hesaplanmasından sonra bütün kümelerin içindeki hataların toplanmasıdır.

- Her küme için rastgele centroid atar. Sonrasında iteratif olarak güncelleme yaparak bu centroidler kümelerin tam ortasına atanır.

- Bu işlemi neye göre yapar; bütün gözlem noktalarının bütün centroidlere olan uzaklığını hesaplar. En yakınları bir küme yapar.

- Bir küme içerisinde verilerimizin centroidlere olan uzaklıklarına göre kümemiz geniş veya dar olabilir.


**Aşamaları**:

- Kaç küme oluşturacağımızı belirtiriz.
- Küme sayısına göre random olarak ilk centroidler atanır.
- Veri gözlemlerinin centroidlere olan uzaklıkları hesaplanarak, bu gözlemleri en yakın oldukları centroidlere ait kümelere atanır.
- Atadığı gözlemlerle centroidler arasındaki mesafeye bakıp centroidlerin konumunu günceller.
- İterasyona devam edip grupları oluşturur.

Farklı değişkenler kullanarak birçok küme analizi yapabiliriz. Her bir küme analizi için en fazla iki değişken kullanırsak, küme etiketlerini daha net bir şekilde tanımlayabiliriz.

İlk olarak, K-Means algoritması bizden küme sayısını (n_clusters) belirlemenizi bekler. Her küme analizi için en uygun küme sayısını çeşitli şekillerde belirleyebiliriz. Bu durumda aşağıda da bahsedeceğimiz Elbow Metodunu kullanacağız.

Son olarak, her analizde farklı bilgiler elde edilir. Bu nedenle her küme analizi sonucunda farklı labellama yapılmalıdır.

Labellama örneği:

**Hopkins Testi**

- Verilerimizin kümelemeye ne kadar yatkın olduğunu gösteren istatistiksel bir testtir.


- Veri, uniform bir şekilde mi dağılmış yoksa rasgele mi dağılmış, bunu belirlemek için kullanılır.


- Veri uniform dağılıma sahipse, yani veriler birbirine benziyorsa, kümeleme yapılamaz. Kümeleme için uygun veriler rasgele dağılmış olmalıdır.


- Hopkins testi iki hipotez ortaya koyar: Null hipotez, verilerin uniform olduğunu (yani rasgele olmadığını) belirtirken, alternatif hipotez verilerin rasgele dağıldığını ifade eder.


- Skor 0.5'in altında olduğunda, yani sıfıra daha yakın olduğunda, verilerin kümeleme için daha uygun olduğu söylenebilir.

```python
# normalde bu fonksiyonumuz  pyclustertend'ın hopkins kütüphanesinde var.Ancak oluşabilecek hatalara karşı 
# doğrudan fonksiyonun kendisini kullanacağız . Bundan dolayı pyclustertend'ı install etmemize gerek yoktur! 
                                                #!pip install pyclustertend --user
                                                # from pyclustertend import hopkins

from sklearn.neighbors import BallTree
import numpy as np
import pandas as pd
def hopkins(data_frame, sampling_size):
    """Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
    no clusterability and a score tending to 0 express a high cluster tendency.
    Parameters
    ----------
    data_frame : numpy array
        The input dataset
    sampling_size : int
        The sampling size which is used to evaluate the number of DataFrame.
    Returns
    ---------------------
    score : float
        The hopkins score of the dataset (between 0 and 1)
    Examples
    --------
    >>> from sklearn import datasets
    >>> from pyclustertend import hopkins
    >>> X = datasets.load_iris().data
    >>> hopkins(X,150)
    0.16
    """
    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)
    # Sample n observations from D : P
    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')
    data_frame_sample = data_frame.sample(n=sampling_size)
    # Get the distance to their neirest neighbors in D : X
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]
    # Randomly simulate n points with the same variation as in D : Q.
    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()
    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)
    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)
    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)
    # Get the distance to their neirest neighbors in D : Y
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist
    # return the hopkins score
    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)
    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')
    return x / (x + y)[0]
```

**Elbow Metodu**

- Optimal küme sayısını belirlemek için kullanılan bir yöntemdir.
- Farklı k değerleri için modelimizi çalıştırırız.
- Her k değeri için datamızdan elde edilecek inertia (SSD) değerleri hesaplanır ve bir grafik çizdirilir.
- Inertia'da en hızlı düşüşün bittiği noktada, yani grafikte bir dirsek oluştuğu noktada, optimal k değeri bulunur.
- Inertia'nın en küçük olduğu nokta, genellikle gözlem sayısı kadar küme demektir ki bu da kümeleme mantığına aykırıdır.


**SSD (Sum of Squared Distances) - Inertia**

- Elbow metoduyla optimal k'nın belirlenmesinde kullanılır.
- Her bir veri noktasının centroidlere olan uzaklıklarının karelerinin toplamı alınarak hesaplanır.
- Inertia'yı minimize etmek hedeflenir. İterasyon, inertia'nın en küçük olduğu noktada durur ve kümeleme tamamlanır.
- Inertia ne kadar küçükse, veri noktalarımızın centroidlere o kadar yakın olduğu anlamına gelir, yani kümeleme daha başarılı olur.

```python
from sklearn.cluster import KMeans

ssd =[]
for n in range(2,11):
    kmeans=KMeans(n_clusters=n,random_state=42)
    kmeans.fit(X1)
    ssd.append(kmeans.inertia_)
plt.figure(figsize=(10,6))
plt.plot(range(2, 11), ssd, "bo-", color='royalblue')
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('ssd')
plt.show()

# Elbow yöntemini kullanırken, dikkat etmemiz gereken bir nokta, range'i 2'den başlatmak ve 1'den başlatmamaktır. 
# Eğer 1'den başlarsak, her zaman optimal k olarak 2'yi göreceğiz.

# Inertia, kümelerdeki gözlemlerin centroidlerini ne sıklıkta çevrelediğini hesaplar. Sadece küme içinde odaklanır.

# Inertia ne kadar küçükse, kümeler kendi içindeki centroidlerini o kadar iyi çevrelemiştir. 
# Bu, kümeleme kalitesinin bir göstergesidir.

# Elbow yöntemi, inertia üzerinden hareket eder ve diğer kümelerle olan uzaklığa bakmaz. 
# Sert bir düşüşün durduğu "ilk" nokta, optimal k olarak belirlenir. Bazı durumlarda, 2-3 tane sert düşüş olabilir.

# Inertia'nın en düşük olduğu nokta, genellikle gözlem sayısı kadar küme anlamına gelir ki 
# bu da kümeleme mantığına aykırıdır. 
# Bu nedenle, genellikle inertia'nın sert düşüşünün durduğu nokta optimal k olarak kabul edilir.

# Bu durumu gözlemlemek için range(2,150) gibi geniş bir aralık kullanabilirsiniz.
```

**Distorsiyon Skoru**

**Silhouette Katsayısı**

- Kümeleme kalitesini değerlendirmek için kullanılır, özellikle labellarımız ve alan bilgimiz yoksa önemlidir.

- İlk olarak, bir küme içindeki bütün gözlemlerin birbirlerine olan uzaklıklarını bulur.

- Sonrasında, bir küme içindeki gözlemlerin diğer kümelerdeki gözlemlere olan uzaklıklarını bulur.

- Her bir gözlem için bir skor hesaplanır ve bu skorlar kullanılarak her küme için ortalama bir skor elde edilir.

- Silhouette katsayısı, 1'e yakınsa iyi bir kümeleme olduğunu gösterir. 

- Amaç, küme içindeki mesafenin düşük, kümeler arası mesafenin ise büyük olmasıdır.

- Inertia küme içi mesafe ile ilgilenirken, Silhouette katsayısı kümeler arası mesafe ile ilgilenir.

```python
# Silhouette skoru, yaptığımız kümelemenin kalitesini ölçen bir metriktir.
# Inertia, her kümenin kendi içerisindeki centroidi etrafında ne sıklıkta kümelenmiş olduğunu ölçer 
# (küme içi mesafelere bakar).
# Ancak, ideal kümelemenin bir diğer önemli ölçütü de maksimum küme arası mesafe idi 
# ve silhouette skoru işte bu durumu hesaba katar.
# Silhouette skoru, 1'e ne kadar yakınsa, kümelemenin o kadar kaliteli olduğunu gösterir. 
# Bir kümenin diğer kümelerden ne kadar ayrıldığını ifade eder.

from sklearn.metrics import silhouette_score
range_n_clusters = range(2,11)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X1)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X1, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
# Yellowbrick'in bu özelliği sayesinde, her bir kümenin ayrı ayrı silhouette skorlarını görebiliyoruz. 
# Silhouette skorlarını hesapladıktan sonra bu skorların görselini çizdiriyoruz. Örneğin, 4 küme için çizdirdik.

# Grafikte, en çok gözlemin ikinci kümede olduğunu görebiliriz. 
# Genişliklerine bakarak, küme içinde az ya da çok gözlem olduğunu anlayabiliriz. Kırmızı çizgi, 
# bütün kümelerimizin silhouette skorlarının ortalamasını temsil eder.

# Cluster'lar arası dağılıma baktığımızda, bütün kümelerin benzer kalitede olduğunu söyleyebiliriz. 
# Bu, kümelememizin homojen ve dengeli olduğunu gösterir.

from sklearn.cluster import KMeans

from yellowbrick.cluster import SilhouetteVisualizer

model3 = KMeans(n_clusters=4, random_state=42)
visualizer = SilhouetteVisualizer(model3)

visualizer.fit(X1)    # Fit the data to the visualizer
visualizer.poof();
```

## Why silhouette_score is negative?

- Varsayalım ki mavi noktalar bir küme ve yeşil noktalar diğer bir küme olarak belirlenmiş olsun. 
- 0 noktasındaki yeni bir gözlemimiz olduğunu farz edelim. Bu gözlem, hangi kümeye atanacağını belirlememiz gerekmektedir.

- Sample'ın hangi kümeye atanacağını yani bu işlemi yaparken, her kümenin merkez noktasını belirleyerek başlarız. Sample hangi kümeye daha yakın ise oraya atanır.
- Mavi noktaların merkez noktasını 1, yeşil noktaların merkez noktasını ise -1'in biraz altında olduğunu varsayalım. Yani mavi noktaların merkezi olan 1'e göre daha yakındır.

- Mavi noktaların merkezi olan 1'e daha yakın olduğu için, yeni gözlem mavi noktaların olduğu kümeye atanır.

- Bir gözlemin hangi kümeyle ilişkilendirileceğini belirlemek için Silhouette Skoru kullanılır. Bu skor, gözlemin kendi kümelerindeki diğer elemanlara olan uzaklığının ortalaması (a) ile en yakın kümedeki elemanlara olan ortalama uzaklık (b) arasındaki farkı ölçer. 

Bu farkı maksimize etmek için, Silhouette Skoru'nu aşağıdaki formülle hesaplarız:



Kırmızı nokta için bu hesaplamayı yapalım:



Bu sonuç, kırmızı noktanın kendi kümesindeki diğer noktalara olan mesafenin, en yakın diğer kümedeki noktalara olan mesafeden daha uzak olduğunu gösterir. 

Yani, bu durumda, kırmızı noktanın kendi kümesinde iyi bir şekilde kümeleşmediği, diğer kümeye ise çok yakın olmadığı anlamına gelir. Bu nedenle, küme sayısını artırmak gerekebilir.

Başka bir deyişle;

- Yeşillerin olduğu küme kendi içlerinde sıkı bir şekilde kümeleşmişken, mavilerin olduğu küme elemanlarının kendi aralarındaki mesafe oldukça fazla.
- Eğer Silhouette skoru (-) değer alıyorsa, o kümedeki sample'lar birbirleri ile iyi kümeleşememiş, aradaki mesafe uzak demektir. Bu durumda kümeleme sayısının artırılması gerekir.
- Silhouette skorunun 0'a yakın veya 0'dan küçük değer alması demek, küme sayısının artırılması gerektiği demektir.

# Hierarchical Clustering

- Biz şu ana kadar Kmeans algoritması üzerinden çıkarımlarımızı yaptık.
- Bir diğer clustering algoritması olan Hierarchical Clustering ile de sonuçları alacağız ve karşılaştıracağız 
- Dikkat etmemiz gereken husus kmeans'teki cluster sayılarımızı göz önünde bulundurcağız ki farklılıkları bulalım

**Hiyerarşik kümeleme**
- Amaç gözlemleri birbirlerine olan benzerliklerine göre kümelerine ayırmaktır.
- Gözlemler daha fazla alt sayıda kümeye ayrılmak istendiğinde kullanılır.
- Hiyerarşik kümeleme yöntemlerinde küme sayısına dendogram sonuçlarına bakarak karar verilirken hiyerarşik olmayan yöntemlerde küme sayısı uygulama yapılmadan önce belirlenir.
- İki metotumuz vardır.

**Agglomerative kümeleme**

- Aşağıdan yukarı - tümevarım. En fazla kullanılan metot.
- Herbir gözleme bakarak benzerliklerine göre birleştire birleştire yukarı çıkar. Sonunda tüm dataya ulaşır.
- Başlangıçta gözlem sayısı kadar küme vardır.
- Veri setinde birbirine en yakın olan iki gözlem bulunur.
- Bu iki nokta bir araya getirilerek yeni bir gözlem oluşturulur. Yani artık veri seti ilk birleşimdeki gözlemlerden oluşmaktadır.
- Aynı işlemler tekrarlanarak yukarı doğru çıkılır. Yani iki kümenin birleşiminden oluşan bu yeni kümeler aynı şekilde birbirlerine benzerliklerine göre tekrar birleştirilir. Bu işlem tüm gözlemler tek bir kümede toplanana kadar tekrar edilir.

**Divisive kümeleme**

- Yukarıdan aşağı - tümdengelim. 
- Yukardan başlayarak tüm datayı sürekli olarak farklılıklarına göre ikiye böle böle kümelemeler birbirinden bağımsız olana kadar iner.
- Başlangıçta tek küme vardır o da veri setidir.
- Tüm gözlemlerin bir arada olduğu küme iki alt kümeye ayrılır.
- Oluşan yeni kümeler birbirlerine benzemeyen alt kümelere bölünür.
- Aynı işlem gözlem sayısı kadar küme elde edilinceye kadar tekrar edilir.

**Not: Karar ağaçlarından farkı; karar ağaçlarında ayırma işlemi hedef değişkene göre yapılırken burada bağımlı değişken olmadığı için gözlemler bağımsız değişkenler üzerinden yapılan uzaklık hesaplarına göre kümelere ayrılır.**

- Hiyerarşik kümelemede başlangıçta küme değerini belirlemek zorunlu değil, ancak KMeans'de zorunlu.
- Ayrıca Hiyerarşik kümelemenin  en büyük dezavantajı predict özelliğinin olmamasıdır. 
- Yeni bir data geldiğinde, datayı veriye ekleyip işlemleri baştan yapmamız gerekiyor.
- Hiyerarşik kümeleme; büyük data setlerinde yavaş olduğundan tercih edilmez.

## Clustering based on Age and Spending Score- X1

```python
from scipy.cluster.hierarchy import linkage

# Dendrogramların oluşturulması sırasında, veriyi ve mesafelerin nasıl ölçüleceğini belirtmek önemlidir.
# Bu nedenle, linkage fonksiyonu içinde, hem veriyi tanımlamamız hem de mesafelerin hangi ölçüt üzerinden hesaplanacağını belirtmemiz gerekir.
# Bu bilgiler, dendrogramın gözlemleri kümelere nasıl atadığını ve bu atama işlemlerini nasıl dallandırdığını belirler.
```

**Hiyerarşik Kümeleme'nin Hyperparametreleri:**

**Affinity:** Yakınlık hiperparametresi, gözlemler arası mesafenin nasıl ölçüleceğini belirler. Varsayılan olarak "euclidean" kullanılır, diğer seçenekler arasında "manhattan", "cosine" ve "precomputed" bulunur. "Ward" kullanıldığında, zorunlu olarak "euclidean" kullanılır.

**Linkage:** Bir gözlemin kümelere nasıl bağlanacağını belirleyen hiperparametredir. Varsayılan olarak "ward" kullanılır, diğer seçenekler arasında "complete", "average" ve "single" bulunur.

**Ward:**
Varyansı düşürecek şekilde gözlemi kümelere bağlar. Bir gözlem, hangi kümenin varyansını daha çok düşürüyorsa o kümeye atanır.


- Varyans, verilerin eşit olduğunu gösteren bir sıfır değerine sahiptir. Varyansın artması, verilerin ortalamadan daha fazla uzaklaştığını ifade eder.
- Veriler ortalama çevresinde kümelenirse, varyans düşük olur. Eğer veriler ortalamadan uzak bir noktaya dağılırsa, varyans yüksek olur.

**Average:**
Her kümenin ortalaması alınır, yeni gelen gözlem hangi kümenin ortalamasını daha çok düşürüyorsa o kümeye atanır.


**Complete:**
Yeni gelen gözlemin her kümenin en uzak gözlemine mesafesi ölçülür, ve bu mesafelerin hangisi daha kısaysa, o kümeye atanır.

**Single:**
Yeni gelen gözlemin her kümenin en yakın gözlemine mesafesi ölçülür, ve bu mesafelerin hangisi daha kısaysa, o kümeye atanır.

```python
# Hiyerarşik Kümeleme için linkage işlemleri
hc_ward = linkage(y=X1, method="ward")    # Varyansa göre atama işlemi
hc_complete = linkage(X1, method="complete")   # En uzak noktaya göre atama işlemi
hc_average = linkage(X1, method="average")     # Ortalamaya göre atama işlemi
hc_single = linkage(X1, method="single")       # En yakın noktaya göre atama işlemi

# Bu 4 farklı durum için, yukarıda oluşturduğumuz değişkenler ile görselleştirmemizi yapacağız
```

```python
from scipy.cluster.hierarchy import dendrogram

# Dendogram, K-means'de kullanılan Elbow yöntemine benzer. 
# Dendogram, verilerin hiyerarşik yapısını görselleştiren bir ağaç diyagramıdır.
# Bu diyagramı oluşturmak için, gözlemler arasındaki mesafeleri ölçen bir linkage metodu kullanılır.
# Veri setimizdeki gözlemler birleşerek bir ağaç oluşturulur.
# Bu agaç, gözlemlerin benzerliklerine ve uzaklıklarına dayanarak dallara ayrılır.
# Dendogram üzerinde kümelerin birbirleri arasındaki mesafeye bakılarak kaç küme seçileceğine karar verilir.
# Dendogram oluşturmak için, verilerin kendisini değil, linkage metodu ile hesaplanan mesafeleri kullanırız.
```

```python
# Burada her bir linkage yöntemi için datamızı kullanarak dendogram çizdiriyoruz. 
# Dendogramı yorumlarken mesafeler dikkate alınır. Kümeler arası mesafe hangisinde fazlaysa onu tercih edeceğiz.
# Küme içi minimum, küme arası maksimum mesafeyi ward metodu sağladığından ward ile devam edeceğiz.
# Dendogramın mavi çizgi ile gösterdiği sayı, bize tavsiye edilen küme sayısıdır.
# Ancak dendogram, kümeler arası mesafeyi maksimum yapacak şekilde bize bir öneride bulunur, 
# ama biz kararımızı sadece dendogramdan değil, scatter plot ve silhouette score'dan da faydalanarak vereceğiz.

# truncate_mode='lastp' ve p=10 ayarlanarak dendogramın sadece son 10 kümeyi bize gösterecek şekilde sadeleştirme yapması 
# sağlanır.

plt.figure(figsize = (20,12))

plt.subplot(221)
plt.title("Ward")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10)

plt.subplot(222)
plt.title("Complete")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_complete, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10)

plt.subplot(223)
plt.title("Average")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_average, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10)

plt.subplot(224)
plt.title("Single")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_single, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10);

# En fazla uzaklığın ward yönteminde olduğunu görüyoruz. Bu yüzden ward'ı seçerek yolumuza devam edeceğiz. 
# Ward, 2 kümeleme yapmamız gerektiğini söylüyor. Fakat 4 kümeleme seçtiğimizde de aradaki mesafelerin hala oldukça uzak
# olduğunu görüyoruz. 
# En uzak mesafe seçilecek diye bir kaide yok. Eğer güzel bir ayrım sağlanabiliyorsa daha fazla kümeleme de yapılabilir.
# Biz kümeleme sayısını 4 olarak belirledik.
```

```python
# Linkage değeri olarak ward'ı belirlediğimizden daha ayrıntılı inceleyebilmek için tekrar çizdiriyoruz.

plt.figure(figsize = (20,8))
plt.title("Dendrogram")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10);
```
